```python
import matplotlib.pylab as plt
import numpy as np

from sigmoid import sigmoid
from step_func import step_function
from ReLU import relu
```

# Step Function

```python
x = np.arange(-5, 5, 0.1)
y = step_function(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

# Sigmoid

```python
x = np.arange(-5, 5, 0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

# 한 번에 표현해보기

```python
x = np.arange(-5, 5, 0.1)
y1 = sigmoid(x)
y2 = step_function(x)

plt.plot(x,y1, label='sigmoid')
plt.plot(x,y2, label='step_func', linestyle='--')
plt.legend()
plt.ylim(-0.1,1.1)
plt.show()
```

# ReLU 함수

```python
x = np.arange(-5, 5, 0.1)
y = relu(x)

plt.plot(x,y)
plt.ylim(-0.1, 5)
plt.show()
```
