# Matplotlib Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 그래프 그리기

```python
x = np.arange(0, 6, 0.1) # 0 ~ 6까지 0.1 간격으로 배열 생성
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle="--",label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("sin&cos")
plt.show()
```
